```python
def plot_dp(dp):
    plt.imshow(dp['img'])
    cs = dp['corners']
    ns = dp['neighs'].numpy()
    plt.plot(cs[:,0].numpy(), cs[:,1].numpy())
    plt.scatter(ns[:,0], ns[:,1])
```

```python
plot_dp(ds[0])
```

```python
for d in ds:
    cs = d['corners']
    ns = d['neighs'].numpy()
    plt.plot(cs[:,0].numpy(), cs[:,1].numpy())
    plt.scatter(ns[:,0], ns[:,1])
```
